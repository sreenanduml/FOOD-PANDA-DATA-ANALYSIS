```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
f["customer_id"]=le.fit_transform(f["customer_id"])
f["gender"]=le.fit_transform(f["gender"])
f["age"]=le.fit_transform(f["age"])
f["city"]=le.fit_transform(f["city"])
f["signup_date"]=le.fit_transform(f["signup_date"])
f["order_id"]=le.fit_transform(f["order_id"])
f["order_date"]=le.fit_transform(f["order_date"])
f["restaurant_name"]=le.fit_transform(f["restaurant_name"])
f["dish_name"]=le.fit_transform(f["dish_name"])
f["category"]=le.fit_transform(f["category"])
f["payment_method"]=le.fit_transform(f["payment_method"])
f["last_order_date"]=le.fit_transform(f["last_order_date"])
f["churned"]=le.fit_transform(f["churned"])
f["rating_date"]=le.fit_transform(f["rating_date"])
f["delivery_status"]=le.fit_transform(f["delivery_status"])
f
```

```python
r.select_dtypes(include='object').columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for col in r:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=r, x=col, order=r[col].value_counts().index)
    plt.title(f'Count Plot: {col}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
r.hist(figsize=(12, 8), bins=20)
plt.suptitle("Histogram of Numerical Columns", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(18,14))
sns.heatmap(f.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=r, x='gender', y='price')
plt.title("Order Price by Gender")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=r, x='churned', y='loyalty_points')
plt.title("Loyalty Points by Churn Status")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=r, x='payment_method', y='price')
plt.title("Order Price by Payment Method")
plt.xticks(rotation=45)
plt.show()
```
